## Creating a Sample DataFrame

```python
q=np.random.randint(1,20,60).reshape(30,2)
p=np.random.randint(500,1000,60).reshape(30,2)
data=np.hstack((q,p))
df=pd.DataFrame(data,columns=['Quantity_A','Quantity_B','Price_A','Price_B'])
df['Revenue_A']=df['Quantity_A']*df['Price_A']
df['Revenue_B']=df['Quantity_B']*df['Price_B']
df.index.names=['Days']
print('As you can see, now the dataframe is showing 6 rows, exactly what we has set above')
df
```

## Plotting the DataFrame

```python
# Pandas Library provides ibuilt plot function. 
# figsize can be passed as an argument that takes a tuple - (width, height)
df.plot(figsize=(11,6)); 
```

## DataFrame columns can be plot separately as well

```python
# subplots=True, plots each column of the dataframe separately. 
# We use semicolon in the end of the plot statement to avoid printing object names of the plot.
df.plot(subplots=True,figsize=(11,10));
```

## Rearranging subplots: layout parameter

```python
# layout parameter (rows,columns)  helps in re-arranging the subplots
# For each of these subplots, the x-axis (days) is shared
df.plot(subplots=True,layout=(3,2),figsize=(13,10));  
plt.tight_layout()
```

## To plot only selected columns of the dataframe 

```python
# in order to plot specific columns, example: Revenue_A & Revenue_B
df[['Revenue_A','Revenue_B']].plot(figsize=(10,4));
plt.tight_layout()
```

## Plotting Histogram of DataFrame

```python
# inbuilt dataframe histogram
# kind depicts the type of graph, width defines the size of each bar,alpha represents the transparency 
# bins defines the number of bars along the dataset
# Note here that both the columns (Series) are plotted on the same graph.
# In order to prevent this, we need to define axes object 

df['A'].plot(kind='hist',width=0.06,bins=10,title='Histogram Comparision',alpha=0.8,figsize=(10,6));
df['B'].hist(width=0.015,bins=10,color='red',figsize=(11,6));
```

## Scatter Plot

```python
# To give color representation, parameter c needs to be defined with the series of the dataframe
df.plot.scatter(x='A',y='B',c='C',cmap='plasma',\
                title='Scatter Plot with Color (c) representing third dimension',figsize=(10,6));
```

```python
# To change the size of each element of the plot wrt to series, 
        # parameter s needs to be initialized with that parameter
df.plot.scatter(x='A',y='B',s=df['C']*100,\
                title='Scatter Plot with Size (s) representing third dimension',figsize=(10,6));
```

## Vertical and Horizontal Bar Graph

```python
fig,axes=plt.subplots(1,2,figsize=(12,5))

x = [5,8,10] 
y = [12,16,6]  
axes[0].bar(x, y, align = 'center')
axes[1].barh(x, y, align = 'center')

x2 = [6,9,11] 
y2 = [6,15,7] 
axes[0].bar(x2, y2, color = 'g', align = 'center',alpha=0.6) # vertical
axes[1].barh(x2, y2, color = 'g', align = 'center',alpha=0.6) # horizontal

axes[0].set_title('Vertical Bar graph')
axes[0].set_ylabel('Y') 
axes[0].set_xlabel('X')

axes[1].set_title('Horizontal Bar graph') 
axes[1].set_ylabel('X') 
axes[1].set_xlabel('Y')

plt.show()
```

```python
x=['A','B','C']
y=[6,8,4]

plt.figure(figsize=(11,6))
plt.title("Allocation")
wedges,texts,autotexts=plt.pie(y,labels=x,explode=[0,0,0.05],autopct='%.2f%%',pctdistance=0.5,labeldistance=0.85)
print('Wedges:',wedges)
print('\nTexts:',texts)
print('\nAutotexts:',autotexts)

# formatting individual components
wedges[2].set(facecolor='pink',edgecolor='brown')
texts[1].set(weight='bold',size=14)
autotexts[1].set(weight='bold',size=14)
```

## Highlighting a single line in Line Plot

```python
# Make a data frame
df=pd.DataFrame({'x': range(1,11), 'y1': np.random.randn(10), 'y2': np.random.randn(10)+range(1,11), 'y3': np.random.randn(10)+range(11,21), 'y4': np.random.randn(10)+range(6,16), 'y5': np.random.randn(10)+range(4,14)+(0,0,0,0,0,0,0,-3,-8,-6), 'y6': np.random.randn(10)+range(2,12), 'y7': np.random.randn(10)+range(5,15), 'y8': np.random.randn(10)+range(4,14) })
 
#plt.style.use('fivethirtyeight')
plt.style.use('seaborn-darkgrid')
my_dpi=96
plt.figure(figsize=(480/my_dpi, 480/my_dpi), dpi=my_dpi)
 
# multiple line plot
for column in df.drop('x', axis=1):
   plt.plot(df['x'], df[column], marker='', color='grey', linewidth=1, alpha=0.4)
 
# Now re do the interesting curve, but biger with distinct color
plt.plot(df['x'], df['y5'], marker='', color='orange', linewidth=4, alpha=0.7)
 
# Change xlim
plt.xlim(0,12)
 
# Let's annotate the plot
num=0
for i in df.values[9][1:]:
   num+=1
   name=list(df)[num]
   if name != 'y5':
      plt.text(10.2, i, name, horizontalalignment='left', size='small', color='grey')
 
# And add a special annotation for the group we are interested in
plt.text(10.2, df.y5.tail(1), 'Mr Orange', horizontalalignment='left', size='small', color='orange')
 
# Add titles
plt.title("Evolution of Mr Orange vs other students", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Time")
plt.ylabel("Score")
```

## Waffle Charts

```python
data = {'Democratic': 48, 'Republican': 46, 'Libertarian': 3}
fig = plt.figure(
    figsize=(11,6),
    FigureClass=Waffle, 
    rows=5, 
    values=data, 
    colors=("#983D3D", "#232066", "#DCB732"),
    title={'label': 'Vote Percentage in 2016 US Presidential Election', 'loc': 'left'},
    labels=["{0} ({1}%)".format(k, v) for k, v in data.items()],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': len(data), 'framealpha': 0}
)
fig.gca().set_facecolor('#EEEEEE')
fig.set_facecolor('#EEEEEE')
plt.show()
```

## Word Clouds

```python
alice_novel = "Lorem Ipsum is simply dummy text of the printing and typesetting industry. Lorem Ipsum has been the industry's standard dummy text ever since the 1500s, when an unknown printer took a galley of type and scrambled it to make a type specimen book. It has survived not only five centuries, but also the leap into electronic typesetting, remaining essentially unchanged. It was popularised in the 1960s with the release of Letraset sheets containing Lorem Ipsum passages, and more recently with desktop publishing software like Aldus PageMaker including versions of Lorem Ipsum."
```

```python
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
#     stopwords=stopwords
)
 
# generate the word cloud
alice_wc.generate(alice_novel)
 
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height
 
# display the cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

There are two types of Plotting styles:
1. Scripting Layes - (procedural method) â€“  using matplotlib.pyplot as â€˜plotâ€™
2. Artist layer (Object oriented method) â€“ using an Axes instance from Matplotlib (preferred)

In the second way, you can use an Axes instance of your current plot and store it in a variable. Example of this is given below

```python
# Making a dummy DataFrame
dict_dummy = {'A':[1, 3, 6, 10, 3], 'B':[3, 7, 70, 60, 4], 'C':[7, 67, 36, 10, 7]}

dummy_df = pd.DataFrame.from_dict(dict_dummy, orient='index').T
```

```python
ax = dummy_df.plot(kind='area', alpha=0.35, figsize=(20, 10))
 
ax.set_title('Graph - 1')
ax.set_ylabel('Random Categories')
ax.set_xlabel('Random Numbers')
```
